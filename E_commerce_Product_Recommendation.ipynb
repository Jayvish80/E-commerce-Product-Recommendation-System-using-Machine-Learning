```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('content_based_recommendation_dataset.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df.reset_index(drop=True, inplace=True)
```

```python
df.columns
```

```python
df['combined_features'] = df['Brand of the product'] + " " + df['Season'] + " " + df['Geographical locations']
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def recommend(product_name):
    idx = df[df['Brand of the product'] == product_name].index[0]

    similarity_scores = list(enumerate(cosine_sim[idx]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    similarity_scores = similarity_scores[1:6]

    product_indices = [i[0] for i in similarity_scores]

    return df['Brand of the product'].iloc[product_indices]
```

```python
recommend(df['Brand of the product'][0])
```

```python
import pickle

pickle.dump(df, open('products.pkl', 'wb'))
pickle.dump(cosine_sim, open('similarity.pkl', 'wb'))
```
